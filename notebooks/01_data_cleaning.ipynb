```python
import pandas as pd
```

## Get Data and DF shapes

```python
df_season_drivers_2023 = pd.read_csv('../data/Raw/Formula1_2023season_drivers.csv')
df_grand_prix_2023 = pd.read_csv('../data/Raw/all_grand_prix_laps_2023.csv')
df_qualifying_results_2023 = pd.read_csv('../data/Raw/Formula1_2023season_qualifyingResults.csv')
df_driver_of_the_day_2023 = pd.read_csv('../data/Raw/Formula1_2023season_driverOfTheDayVotes.csv')

print(df_season_drivers_2023.shape)
print(df_qualifying_results_2023.shape)
print(df_grand_prix_2023.shape)
print(df_driver_of_the_day_2023.shape)
```

## Convert Qualifying time to seconds

```python
df_qualifying_results_2023_fl = df_qualifying_results_2023.dropna()
df_qualifying_results_2023_fl = df_qualifying_results_2023_fl.replace("DNF", '0.00')
df_qualifying_results_2023_fl["Q3"] = df_qualifying_results_2023_fl["Q3"].astype(str).str.split(':').apply(lambda x:  0.00 if (len(x)) < 2 else (float(x[0]) * 60) + float(x[1]))
df_qualifying_results_2023_fl["Q1"] = df_qualifying_results_2023_fl["Q1"].str.split(":").apply(lambda x: (float(x[0]) * 60) + float(x[1]))
df_qualifying_results_2023_fl["Q2"] = df_qualifying_results_2023_fl["Q2"].str.split(":").apply(lambda x: (float(x[0]) * 60) + float(x[1]))
df_qualifying_results_2023_fl = df_qualifying_results_2023_fl.dropna()
```
